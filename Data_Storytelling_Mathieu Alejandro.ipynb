**Respondiendo la pregunta 4**

```python
%cd '/content/gdrive/MyDrive/DataScience/Desafiodatascience/'
df_games = pd.read_csv ('Video_Games.csv',',')
sales_by_publisher = df_games.groupby('Publisher')['Global_Sales'].sum().reset_index()
top_10_publishers = sales_by_publisher.nlargest(10, 'Global_Sales')
other_sales = sales_by_publisher[~sales_by_publisher['Publisher'].isin(top_10_publishers['Publisher'])]['Global_Sales'].sum()
other_row = pd.DataFrame({'Publisher': ['Otros'], 'Global_Sales': [other_sales]})
final_df = pd.concat([top_10_publishers, other_row])
plt.figure(figsize=(10, 6))
sns.barplot(x='Publisher', y='Global_Sales', data=final_df, palette='viridis')
plt.title('Top 10 Publishers según Ventas Globales')
plt.xlabel('Publisher')
plt.ylabel('Ventas Globales (en millones)')
plt.xticks(rotation=45)
plt.show()
```


Mediante este gráfico podemos agrupar los 10 mejores Publisher en términos de ventas globales, como así también los que no están en el top 10, creandose una nueva categoría denominada "otros".

```python

genre_rating_cross = pd.crosstab(df_games['Genre'], df_games['Rating'])

top_ratings = genre_rating_cross.apply(lambda row: row.nlargest(2).index.tolist(), axis=1)

num_plots = len(genre_rating_cross)

num_rows = (num_plots // 6) + (num_plots % 6)
num_cols = 6

fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(12, 4*num_rows))

for idx, (genre, ax) in enumerate(zip(genre_rating_cross.index, axes.flatten())):
    top_ratings_for_genre = top_ratings.loc[genre]
    genre_rating_cross.loc[genre, top_ratings_for_genre].plot(kind='pie', autopct='%1.1f%%', startangle=140, cmap='Set3', ax=ax)
    ax.set_title(f'{genre}')
    ax.set_ylabel('')

plt.tight_layout()
plt.show()
```

**Calculo del R² (R cuadrado) - verificando las predicciones del modelo.**

El puntaje R² (R cuadrado) es una medida de cuánto se ajustan las predicciones del modelo a los valores reales. Su valor varía entre 0 y 1, donde 1 indica un ajuste perfecto del modelo a los datos y 0 indica que el modelo no explica nada de la variabilidad de los datos.

R² Score en el conjunto de entrenamiento (0.9953): Esto indica que el modelo explica aproximadamente el 99.53% de la variabilidad de los datos en el conjunto de entrenamiento. Un valor tan alto sugiere que el modelo está haciendo un excelente trabajo al ajustarse a los datos de entrenamiento.

R² Score en el conjunto de prueba (0.8074): Esto indica que el modelo explica aproximadamente el 80.74% de la variabilidad de los datos en el conjunto de prueba. Un valor del 80% también es bastante bueno, lo que sugiere que el modelo generaliza bien a datos no vistos (el conjunto de prueba).

En resumen, el modelo de Random Forest Regressor parece estar funcionando muy bien. En este caso, un R² de alrededor del 80% en el conjunto de prueba indica que el modelo está generalizando bien y puede ser útil para hacer predicciones sobre los datos del modelo y que son bastante cercanas a los valores reales de las ventas globales de videojuegos.

**Calculo del Error Cuadrático Medio (MSE)**

```python
# Realizar predicciones en el conjunto de prueba
y_pred = rf_model.predict(X_test)

# Importar la métrica MSE
from sklearn.metrics import mean_squared_error

# Calcular el MSE en el conjunto de prueba
mse_test = mean_squared_error(y_test, y_pred)

# Mostrar el resultado
print("Error cuadrático medio (MSE) en el conjunto de prueba:", mse_test)
```

En este caso, un MSE de aproximadamente 0.795 significa que, en promedio, las predicciones del modelo están desviadas en 0.795 unidades cuadradas del valor real.

En términos generales, cuanto más bajo sea el MSE, mejor será el rendimiento del modelo, ya que indica que las predicciones están más cerca de los valores reales.

Un MSE de 0.795 no es malo, pero tampoco es excelente. Para evaluar si es bueno o malo, se deben considerar el contexto del problema y compararlo con otros modelos o métodos.
